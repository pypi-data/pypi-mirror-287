```python
# only get clusters
clusters = []
for ic in initial_clusters:
    if len(ic)>1:
        clusters.append(ic)

# transfer ids to aggregate_id
ids = {}
for ni in news_item_list:
    ids[ni['news_items'][0]['id']] = ni['id']

clusters_correct_id = []
for c in clusters:
    clusters_correct_id.append([ids.get(id) for id in c])

# deep copy list as pop is used in grouping algo
all_clusters_correct_id = copy.deepcopy(clusters_correct_id)
```

```python
# use grouping algo
results = [group_aggregate(aggregate_ids, news_item_list) for aggregate_ids in clusters_correct_id]

clusters_only = copy.deepcopy(results)
```

```python
# ad non-clusters
flat_list = list(itertools.chain(*all_clusters_correct_id))

no_clusters = []
for ni in news_item_list:
    if ni['id'] not in flat_list:
        no_clusters.append(ni['id'] )
        results.append(ni)
```

```python
# save result to json

with open("7days_20231120_clusters.json", "w", encoding="utf-8") as output_file:
    json.dump(results, output_file)
```
